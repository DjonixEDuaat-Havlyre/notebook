## New install
---

[] Go to target directory

```
mkdir <new_dir_name>
cd <new_dir_name>
```

[] Create Virtual Env

```
conda create -n <env_name>[ <package_name>[=<package_version>] ...]
```

`conda create -n dj django`

[] Create Project

```
django-admin startproject <site_name>
```

[] Test Site Install

```
python manage.py runserver[ <ipaddress>:<port>]
```

`python manage.py runserver 127.0.0.1:8080`

[] Create App

```
cd <site_name>
python manage.py startapp <app_name>
```

````python
####### models.py 

from django.db import models

class CareLevel(models.Model):
    '''Code indicating care level'''
    carelevel_text = models.CharField(max_length=5)
    description = models.CharField(max_length=100)


class Client(models.Model):
    '''Client specific data about a person'''
    person = models.ForeignKey(Person, on_delete=models.CASCADE)
    carelevel = models.ForeignKey(CareLevel)
    

class Employee(models.Model):
    '''Employee specific data about a person'''
    person = models.ForeignKey(Person, on_delete=models.CASCADE)
    role = models.ForeignKey(Role)
    

class Role(models.Model):
    '''A job code'''
    role_text = models.CharField(max_length=10)
    description = description = models.CharField(max_length=100)
    

class Person(models.Model):
    '''A person to enter in the db'''
    fullname = models.ForeignKey(FullName, on_delete=models.CASCADE)
    pub_date = models.DateTimeField('date published')


class FullName(models.Model):
    '''I may have to import .Name'''
    given_name = Name(name_type=False)
    ancestoral_name = Name(name_type=False)
    family_name = Name(name_type=True)
    
    
class Name(models.Model):
    '''An atomic Name; name_type indicates first, last, etc'''
    name_text = models.CharField(max_length=35)
    # boolean name_type --> false for given or ancestoral, true for family
    name_type = models.Boolean()
    
````
